```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# navigating to File path

csv_path = "C:\\Users\\skavy\\Desktop\\Bootcamp\\Pandas\\HeroesOfPymoli\\purchase_data.csv"

# Reads the file into Pandas data frame

purchase_data = pd.read_csv(csv_path)
```

```python
#Gets Total number of unique players

Players = purchase_data["SN"].nunique()
Total_Players = pd.DataFrame({"Total Players": [Players]})
Total_Players
```

```python
#PURCHASE ANALYSIS (TOTAL)
#-------------------------------------------------------
#Gets number of unique items.
Unique_Items = len(purchase_data["Item ID"].unique())

# Gets Average Price.
Average_Price = "$" + str(np.around(purchase_data["Price"].mean(),2))

#Gets number of Purchases.
Purchases_Count = purchase_data["Purchase ID"].count()

#Gets total revenue.
Revenue = "$" + str(np.around(purchase_data["Price"].sum(),2))

#Reading purchase Analysis data into a new dataframe.
Purchasing_Summary = pd.DataFrame({"Number Of Unique Items": [Unique_Items], "Average Price": [Average_Price], \
                                   "Number Of Purchases": [Purchases_Count], "Total Revenue": [Revenue]})

Purchasing_Summary
```

```python
#GENDER DEMOGRAPHICS
#---------------------------------------------------------
# Grouped purchase_data by Gender.
Gender_Count = purchase_data.groupby("Gender")

# Counts the total unique names in "SN" column by gender.
Unique_Gender_Count = Gender_Count.nunique()["SN"]

#Percentage and Count of unique gender players.
Gender_Percent = (purchase_data["Gender"].value_counts(normalize=True) * 100)
Gender_Percent

#Reads Gender Demographics into a new DataFrame.
Gender_Demo = pd.DataFrame({"Total Count": Unique_Gender_Count,"Percentage of Players": Gender_Percent.map("${:,.2f}".format)})
Gender_Demo[['Total Count', 'Percentage of Players']].sort_values('Total Count', ascending=False).head()
```

```python
#PURCHASING ANALYSIS(GENDER)
#----------------------------------------------------------

#Gets purchase count by gender
purchase_by_gender = Gender_Count["Purchase ID"].count()
purchase_by_gender

# Gets Average purchase prices by gender
Avg_price = Gender_Count["Price"].mean()
Avg_price

# Gets Average purchase total by gender 
Total_purchase = Gender_Count["Price"].sum()
Total_purchase

# Gets Average purchase total per person by gender
Avg_Per_Person = (Total_purchase/ Unique_Gender_Count)
Avg_Per_Person

#Purchasing Analysis (Gender)
Gender_Analysis = pd.DataFrame({"Purchase Count": purchase_by_gender,"Average Purchase Price": Avg_price.map("${:,.2f}".format), \
                                "Total Purchase Value": Total_purchase.map("${:,.2f}".format),"Avg Total Purchase per Person": Avg_Per_Person.map("${:,.2f}".format)})

Gender_Analysis
```

```python
#AGE DEMOGRAPHICS
#--------------------------------------------------
#Creates bins for ages
bins = [0,9,14,19,24,29,34,39,150]

#creates labels for age groups
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorizes the existing players using the age bins
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_labels)

#Gets the numbers and percentages by age group
Age_GroupedBy = purchase_data.groupby("Age Group")
Age_Group_Total =Age_GroupedBy["SN"].nunique()
Age_Group_Sum = Age_Group_Total.sum()
Age_Group_Percent = (Age_Group_Total / Age_Group_Sum) * 100

#Creates a new summary data frame to hold the above results of the cell and displays.
Age_Demo = pd.DataFrame({"Total Count": Age_Group_Total, "Percentage of Players": Age_Group_Percent.map("{:,.2f}%".format)})
Age_Demo.reset_index(inplace=True)

Age_Demo
```

```python
#TOP SPENDERS
#------------------------------------------------------------------------ 
#Run basic calculations to obtain the results in the table below
SN_GroupedBy = purchase_data.groupby("SN")
Spender_Purchase = SN_GroupedBy["Purchase ID"].count()
Spender_Avg_Price = SN_GroupedBy["Price"].mean()
Spender_Total = (SN_GroupedBy["Price"].sum())

#Create a summary data frame to hold the results
Spender_df = pd.DataFrame({"Purchase Count": Spender_Purchase,"Average Purchase Price":Spender_Avg_Price.map("${:,.2f}".format), \
                           "Total Purchase Value": Spender_Total})
Spender_df.head()

#Sort the total purchase value column in descending order
Spender_Sorted = Spender_df.sort_values(["Total Purchase Value"],ascending=False)

#Optional: give the displayed data cleaner formatting


#Display a preview of the summary data frame
Spender_Sorted.reset_index(inplace=True)
Spender_Sorted.head()

```

```python
#Most Popular Items
#Retrieve the Item ID, Item Name, and Item Price columns
Items_Col = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
Items_df = Items_Col.rename(columns ={"Price": "Item Price"})
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
Item_grouped = Items_df.groupby(["Item ID","Item Name"])

#purchase count
Item_purchase_count = Item_grouped["Item ID"].count()

#item price
Item_price = Item_grouped["Item Price"].mean()

#total purchase value
Total_value = Item_grouped["Item Price"].sum()

#Create a summary data frame to hold the results
Most_Popular_Items = pd.DataFrame({"Purchase Count":Item_purchase_count,"Item Price":Item_price.map("${:,.2f}".format), \
                                   "Total Purchase Value":Total_value})

#Sort the purchase count column in descending order
Sorted_Purchase_Count = Most_Popular_Items.sort_values(by="Purchase Count", ascending=False)

#Optional: give the displayed data cleaner formatting


#Display a preview of the summary data frame
Sorted_Purchase_Count["Total Purchase Value"].map("${:,.2f}".format).head()
```

```python
#Most Profitable Items
# --------------------------------------------------------------------------------- 
#Sort the above table by total purchase value in descending order
Sorted_Purchase_Value = Most_Popular_Items.sort_values(by="Total Purchase Value", ascending=False)

#Optional: give the displayed data cleaner formatting


#Display a preview of the data frame
Sorted_Purchase_Value["Total Purchase Value"].map("${:,.2f}".format).head().head()
```
