```python
#Defining display table preferences

# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "C:\\Users\\skavy\\Desktop\\Bootcamp\\Pandas\\PyCitySchools\\schools_complete.csv"
student_data_to_load = "C:\\Users\\skavy\\Desktop\\Bootcamp\\Pandas\\PyCitySchools\\students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#DISTRICT SUMMARY
#-------------------------------------------------------------------------
#Calculate the total number of schools
Schools_Count = school_data_complete["school_name"].nunique()

#Calculate the total number of students
Students_Count = school_data_complete["student_name"].count()

#Calculate the total budget
Budget = school_data_complete["budget"].unique()
Total_Budget = "$" + str(np.around(Budget.sum(),2))

#Calculate the average math score 
Avg_Math_Score = np.around(school_data_complete["math_score"].mean(),6)

#Calculate the average reading score
Avg_Reading_Score = np.around(school_data_complete["reading_score"].mean(),5)

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Overall_Rate = np.around(((Avg_Math_Score + Avg_Reading_Score)/2),6)

#Calculate the percentage of students with a passing math score (70 or greater)
Math_pass = school_data_complete.loc[school_data_complete["math_score"] > 69, :]
Math_pass_percent = np.around((len(Math_pass)/ len(school_data_complete["math_score"])) * 100,6)


#Calculate the percentage of students with a passing reading score (70 or greater)
Reading_pass = school_data_complete.loc[school_data_complete["reading_score"] > 69, :]
Reading_pass_percent = np.around((len(Reading_pass) / len(school_data_complete["reading_score"])) * 100,6)

#Create a dataframe to hold the above results
District_Summary = pd.DataFrame({"Total Schools": [Schools_Count],"Total Students": [Students_Count], \
                                 "Total Budget": [Total_Budget],"Average Math Score":[Avg_Math_Score], \
                                 "Average Reading Score":[Avg_Reading_Score], "% Passing Math": [Math_pass_percent], \
                                 "% Passing Reading": [Reading_pass_percent],"% Overall Passing Rate": [Overall_Rate]
                                })
District_Summary
```

```python
#SCHOOL SUMMARY
#----------------------------------------------------------------------------------
#School Name
School_Name = school_data_complete.groupby(['school_name'])

#School Type
School_Type = School_Name["type"].first()

#Total Students
Total_Students = School_Name["Student ID"].count()

#Total School Budget
Total_Sch_Budget = School_Name["budget"].first()
#Per Student Budget
Per_Stu_Budget = Total_Sch_Budget / Total_Students

#Average Math Score
School_Avg_math = School_Name["math_score"].mean()

#Average Reading Score
School_Avg_read = School_Name["reading_score"].mean()

#% Passing Math
School_mathpass_percent= (school_data_complete[school_data_complete['math_score'] >= 70] \
                                             .groupby(["school_name"])['Student ID'].count()/Total_Students ) * 100

#% Passing Reading
School_readpass_percent= (school_data_complete[school_data_complete['reading_score'] >= 70] \
                                             .groupby(["school_name"])['Student ID'].count()/Total_Students ) * 100

#Overall Passing Rate (Average of the above two)
School_overallpass_percent = (School_mathpass_percent + School_readpass_percent) / 2

#Create a dataframe to hold the above results
School_Summary = pd.DataFrame({"School Type": School_Type,"Total Students":Total_Students, \
                               "Total School Budget": Total_Sch_Budget,"Per Student Budget": Per_Stu_Budget, \
                               "Average Math Score": School_Avg_math, "Average Reading Score":School_Avg_read, \
                               "% Passing Math":School_mathpass_percent, "% Passing Reading": School_readpass_percent, \
                               "% Overall Passing Rate": School_overallpass_percent})

# TOP PERFORMING SCHOOLS (By Passing Rate)
#-------------------------------------------------------------------------------------
top_5 = School_Summary.sort_values("% Overall Passing Rate", ascending = False)
top_5.head()
```

```python
# BOTTOM PERFORMING SCHOOLS (By Passing Rate)
#-------------------------------------------------------------------------------------
bottom_5 = School_Summary.sort_values("% Overall Passing Rate")
bottom_5.head()
```

```python
# MATH SCORES BY GRADE
# -----------------------------------------------------------------------------------------
 
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement(using .loc)
#Groups each series by school(using Groupby)
Math_Avg_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"] \
               .groupby("school_name")["math_score"].mean()
Math_Avg_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"] \
                .groupby("school_name")["math_score"].mean()
Math_Avg_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"] \
                .groupby("school_name")["math_score"].mean()
Math_Avg_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"] \
                .groupby("school_name")["math_score"].mean()

#Combine the series into a dataframe
Math_by_grade = pd.DataFrame({"9th":Math_Avg_9th,"10th": Math_Avg_10th, "11th": Math_Avg_11th, "12th": Math_Avg_12th })
Math_by_grade
#Optional: give the displayed data cleaner formatting
```

```python
# READING SCORES BY GRADE
# --------------------------------------------------------------------------------------------------
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
Reading_Avg_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"] \
                  .groupby("school_name")["reading_score"].mean()
Reading_Avg_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"] \
                   .groupby("school_name")["reading_score"].mean()
Reading_Avg_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"] \
                   .groupby("school_name")["reading_score"].mean()
Reading_Avg_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"] \
                   .groupby("school_name")["reading_score"].mean()

#Combine the series into a dataframe
Reading_by_grade = pd.DataFrame({"9th":Reading_Avg_9th,"10th": Reading_Avg_10th, \
                                 "11th": Reading_Avg_11th, "12th": Reading_Avg_12th})
Reading_by_grade
#Optional: give the displayed data cleaner formatting
```

```python
# SCORES BY SCHOOL SPENDING
# -------------------------------------------------------------------------------------------
 #Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
spending_bins = [0, 585, 615, 645, 675]
group_names = ["(Low) <$585", "(Med 1) $585-615", "(Med 2) $615-645", "(High) $645-675"]


School_Summary["Spending Ranges(Per Student)"] = pd.cut(Per_Stu_Budget, spending_bins, labels=group_names)
#Average Math Score
Math_Spending = School_Summary.groupby(["Spending Ranges(Per Student)"])["Average Math Score"].mean()

#Average Reading Score
Reading_Spending = School_Summary.groupby("Spending Ranges(Per Student)")["Average Reading Score"].mean()

#% Passing Math
Mathpass_Spending = School_Summary.groupby("Spending Ranges(Per Student)")["% Passing Math"].mean()

#% Passing Reading
Readingpass_Spending = School_Summary.groupby("Spending Ranges(Per Student)")["% Passing Reading"].mean()

#Overall Passing Rate (Average of the above two)
Overall_Spending = (Mathpass_Spending + Readingpass_Spending) / 2

# creates a new frame holding the above data
Scores_By_School_Spending = pd.DataFrame({"Average Math Score": Math_Spending,"Average Reading Score": Reading_Spending, \
                                          "% Passing Math": Mathpass_Spending,"% Passing Reading": Readingpass_Spending, \
                                          "Overall Passing Rate": Overall_Spending})
                                                  
                                                   
Scores_By_School_Spending

```

```python
# SCORES BY SCHOOL SIZE
# ------------------------------------------------------------------------------------
# Bin and label group

size_bins = [0, 1000, 2000, 5000]
Size_group = ["< 1000", "1000 - 2000", "2000 - 5000"]

School_Summary["School Size"] = pd.cut(School_Summary["Total Students"], size_bins, labels=Size_group)

#Average Math Score
Math_by_schoolsize = School_Summary.groupby(["School Size"])["Average Math Score"].mean()

#Average Reading Score
Reading_by_schoolsize = School_Summary.groupby(["School Size"])["Average Reading Score"].mean()

#% Passing Math
Mathpass_by_schoolsize = School_Summary.groupby(["School Size"])["% Passing Math"].mean()

#% Passing Reading
Readingpass_by_schoolsize = School_Summary.groupby(["School Size"])["% Passing Reading"].mean()

#Overall Passing Rate (Average of the above two)
Overall_by_schoolsize = (Mathpass_by_schoolsize + Readingpass_by_schoolsize) / 2

# creates a new frame holding the above data
Scores_By_School_Size = pd.DataFrame({"Average Math Score": Math_by_schoolsize,"Average Reading Score": Reading_by_schoolsize, \
                                      "% Passing Math": Mathpass_by_schoolsize,"% Passing Reading": Readingpass_by_schoolsize, \
                                      "Overall Passing Rate": Overall_by_schoolsize})
                                                  
                                                   
Scores_By_School_Size
```
